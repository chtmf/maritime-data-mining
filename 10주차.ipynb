```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
from scipy import stats

df = pd.read_csv('branch_satisfaction.csv')

print('[데이터 미리보기]')
print(df.head(), '\n')

a = df[df['branch'] == 'A']['score']
b = df[df['branch'] == 'B']['score']
c = df[df['branch'] == 'C']['score']

result = stats.f_oneway(a, b, c)
means = df.groupby('branch')['score'].mean().sort_values(ascending=False)

print('[분석 결과]')
print('지점별 평균')
print(means, '\n')
print('F-statistic:', round(result.statistic, 4))
print('p-value:', round(result.pvalue, 6))
print('평균이 가장 높은 지점:', means.index[0])

print('\n[결과 해석]')
if result.pvalue < 0.05:
    print('지점 간 평균 고객만족도 차이가 유의합니다.')
else:
    print('지점 간 평균 고객만족도 차이가 유의하지 않습니다.')
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
```

```python
file_path = 'ad_age_sales.csv'
```

```python
df = pd.read_csv(file_path)

print('[데이터 미리보기]')
print(df.head(), '\n')
```

```python
model = smf.ols(
    'sales_amount ~ C(ad_type) + C(age_group) + C(ad_type):C(age_group)',
    data=df
).fit()

anova_table = anova_lm(model, typ=2)
```

```python
print('[ANOVA 결과표]')
print(anova_table, '\n')

p_ad = anova_table.loc['C(ad_type)', 'PR(>F)']
p_age = anova_table.loc['C(age_group)', 'PR(>F)']
p_inter = anova_table.loc['C(ad_type):C(age_group)', 'PR(>F)']

print('[핵심 p-value]')
print('광고유형 p-value:', p_ad)
print('연령대 p-value:', p_age)
print('상호작용 p-value:', p_inter)
```

```python
print('\n[결과 해석]')
if p_inter < 0.05:
    print('상호작용 효과가 유의 → 광고 효과는 연령대에 따라 달라짐')
else:
    print('상호작용 효과 없음 → 주효과 중심 해석')
```
